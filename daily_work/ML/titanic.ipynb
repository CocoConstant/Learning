```python
import matplotlib.pyplot as plt

label = ['lived', 'died']
lived = data[data['Survived'] == 1].Sex.value_counts().to_dict()
died = data[data['Survived'] == 0].Sex.value_counts().to_dict()
situation = {'lived':lived, 'died':died}

fig, ax = plt.subplot()
for sit, tag in situation.items():
    p = ax.bar(['lived', 'died'], tag.values(), width=0.5, label=['male', 'female'])
```
