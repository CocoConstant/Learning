```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plot_data = np.array([[0.9698, 0, 0], 
                      [0, 0.9961, 0], 
                      [0, 0, 0.9392]])
```

```python
# 使用 seaborn 的 heatmap 函数绘制热图
plt.figure(figsize=(5, 4))  # 设置图形的大小
sns.heatmap(plot_data, annot=True, fmt=".2%", cmap="YlGnBu", cbar=True, 
            annot_kws={"size": 12}, linewidths=0.5, linecolor="white")

# 设置标题和标签
plt.title('Accuracy of model prediction')
plt.xticks(ticks=[0.5, 1.5, 2.5], labels=['GSE240252', 'GSE266577', 'GSE222649'], rotation=45)
plt.yticks(ticks=[0.5, 1.5, 2.5], labels=['GSE240252', 'GSE266577', 'GSE222649'], rotation=360)

# 显示图形
plt.show()
```
