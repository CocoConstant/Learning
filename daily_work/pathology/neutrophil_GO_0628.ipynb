```python
import gseapy as gp
import matplotlib.pyplot as plt
import pandas as pd
```

```python
deg_data = pd.read_csv("/mnt/c/Users/ouyangkang/Downloads/neu_not_remove.csv")
```

```python
up_data = deg_data[deg_data.logFC > 0]
down_data = deg_data[deg_data.logFC < 0]
```

```python
up_data = up_data.sort_values("pvals_adj")
up_data
```

```python
down_data = down_data.sort_values("pvals_adj")
down_data
```

```python
up_gene_list = up_data.BC_1.tolist()[0:99]
down_gene_list = down_data.BC_1.tolist()[0:99]
```

```python
enr = gp.enrichr(gene_list=down_gene_list,
                 gene_sets=["GO_Biological_Process_2023"],
                 organism="human",
                 outdir=None)
enr.results
```

```python
from gseapy import barplot, dotplot

ax = barplot(enr.results,
              column="Adjusted P-value",
              group='Gene_set', # set group, so you could do a multi-sample/library comparsion
              size=10,
              top_term=20,
              figsize=(6, 10),
              #color=['darkred', 'darkblue', "pink"] # set colors for group
              #color = {'KEGG_2021_Human': 'salmon', 'MSigDB_Hallmark_2020':'darkblue'}
             )
```

```python
enr.results[enr.results["Adjusted P-value"] < 0.05]
```
