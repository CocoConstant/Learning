```python
import pandas as pd
import numpy as np
```

```python
dir_path = "/mnt/d/dailywork/01_Smear_seq/cancer/eso_squamou_cancer/geo/GSE53625/"
# get sample annotation infomation
sample_info = {}
with open(dir_path + "GSE53625_series_matrix.txt", "r") as f:
    for line in f:
        if line.startswith("!Sample_"):
            key, value = line.strip().split("\t", 1)
            key = key.strip("!")
            sample_info[key] = value.split("\t")
sample_info_df = pd.DataFrame(sample_info)

# get sample*gene matrix
# find start row of sample*gene matrix
start_line = 0
with open(dir_path + "GSE53625_series_matrix.txt", "r") as f:
    for idx, line in enumerate(f):
        if line.startswith("!series_matrix_table_begin"):
            start_line = idx + 1
            break
matrix_df = pd.read_csv(dir_path + "GSE53625_series_matrix.txt", sep="\t", skiprows=start_line, comment="!")
```

```python
sample_info_df = sample_info_df[['Sample_title', 'Sample_geo_accession']]
sample_info_df.head()
```

```python
sample_info_df["Sample_type"] = [i.strip("\"").strip() for i in sample_info_df.Sample_title.str.split("from").str[0]]
sample_info_df["Patient_ID"] = [i.strip("\"") for i in sample_info_df.Sample_title.str.split("from patient").str[1]]
sample_info_df['Sample_geo_accession'] = [i.strip("\"") for i in sample_info_df['Sample_geo_accession']]
sample_info_df.head()
```

```python
sample_info_df[["Patient_ID", "Sample_type", "Sample_geo_accession"]].to_csv(dir_path + "patient_geo_accession.csv", index=None)
```

```python
probe2gene = pd.read_csv("/mnt/d/dailywork/01_Smear_seq/cancer/eso_squamou_cancer/geo/probe2gene_GPL19748.csv", index_col=0)
probe2gene.head()
```

```python
start_index = 0
end_index = 0
with open("/mnt/d/dailywork/01_Smear_seq/cancer/eso_squamou_cancer/geo/GSE53625/GSE53625_family.soft", 'r') as f:
    for idx, line in enumerate(f):
        if line.startswith('!platform_table_begin'):
            start_index = idx + 1
        if line.startswith('!platform_table_end'):
            end_index = idx -1
            break

id2probe = pd.read_csv("/mnt/d/dailywork/01_Smear_seq/cancer/eso_squamou_cancer/geo/GSE53625/GSE53625_family.soft", sep='\t', skiprows=start_index, nrows=end_index - start_index)
print(id2probe.shape)
id2probe.head()
```

```python
probe2gene.rename(columns={"probe_id": "SPOT_ID"}, inplace=True)
```

```python
id_symbol_info = id2probe.merge(probe2gene, on="SPOT_ID")
print(id_symbol_info.shape)
id_symbol_info.head()
```

```python
id_symbol_info.drop_duplicates(subset=['ID'], inplace=True)
```

```python
id_symbol_dict = dict(zip(id_symbol_info['ID'], id_symbol_info['symbol']))
```

```python
matrix_df["ID_REF"] = matrix_df["ID_REF"].apply(lambda x: id_symbol_dict.get(x, np.nan))
```

```python
matrix_df.dropna(inplace=True)
```

```python
matrix_df.rename(columns={'ID_REF': 'Gene'}, inplace=True)
```
