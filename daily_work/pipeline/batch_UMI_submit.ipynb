```python

for num, fq in enumerate(data.FASTQ):
    final_list = []
    fq_list = [i.strip().strip('\'') for i in fq.strip('[]').split(',')]
    fq1_list = [i for i in fq_list if '1.fq.gz' in i]
    fq2_list = [i for i in fq_list if '2.fq.gz' in i]

    for i in fq1_list:
        judge_str = i[0:-8]
        paired_fq = [k for k in fq2_list if k[0:-8] == judge_str][0]
        final_list.append([i, paired_fq])
    
    final_list = str(final_list)
    final_list = final_list.replace('\'', '')
    data['FASTQ'].iloc[num] = final_list
```
